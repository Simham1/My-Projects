So by observation, we can observe that in the match_df we had taken first innings total runs (target) and in second case, that is in the delivery_df we consider inning two runs, as our main aim is to find the probability of either team to win or loose. we need current runs and runrate,
so for current runs, we can apply groupby on match_id and take the cummulative sum wrt total_runs_y. Now, basically total_runs was present in the match_df as well as delivery_df, but as we merge both the dataframe, it resulted total_runs_x and total_runs_y, so total_runs_x is the first innings runs + 1 (target) and total_runs_y are the second innings runs, ball by ball, by applying cummulative_sum, this becomes current score.

total_runs_y gives the run scored after each ball, so in the second innings, we want to get the total second innings runs, so we will groupby match_id and will apply the commulative sum

```python
delivery_df['total_runs_y'] = delivery_df['total_runs_y'].astype(int)
```

```python
print(delivery_df['total_runs_y'].dtype)
print(delivery_df['total_runs_y'].unique())
```

```python
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'], errors='coerce')
```

```python
delivery_df['current_score'] = delivery_df['total_runs_y'].cumsum()
delivery_df
```

```python
# runs left
delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']
delivery_df
```

```python
# balls left
# balls left
"""
if one ball is played, then balls left = 120-1 = 119 .............case 1
similarly, 120-2 = 118, 120-3 = 117....

so similarly, if over = 1, over has 6 balls, right, so 1*6 = 6
now, ball = 1, so 6+1 = 7, now = 126 -7 = 119, which is same as (case 1)

so we'll use ball_left = 126 - (over*6 + current_ball) ***************************************
......

"""

delivery_df['balls_left'] = 126-(delivery_df['over']*6+delivery_df['ball'])
delivery_df.head(20)
```

```python
delivery_df['player_dismissed'].unique()
```

```python
# filling nan values with zero

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')
# now we will convert this player_dismissed col into a boolean col
# if the player is not dismissed then it's 0 else its 1
delivery_df['player_dismissed']= delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else '1')

#Convert String to int
delivery_df['players_dismissed'] = delivery_df['player_dismissed'].astype(int)


delivery_df['players_dismissed'].unique()
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(pd.to_numeric, errors='coerce').astype(int)
```

```python
delivery_df['player_dismissed'].dtypes
```

```python
# wickets fallen

wickets = delivery_df['player_dismissed'].cumsum().values

delivery_df['wickets'] = 10-wickets

delivery_df
```

```python
wickets
```

```python
# current run rate

delivery_df['current_run_rate'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])

# Required run rate

delivery_df['req_run_rate'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=1,stratify=test)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
# batting_team, bowling_team and city are categorical columns/variables
# they will be converted to numeric using onehot encoder method

cf = ColumnTransformer([
    ('trf', OneHotEncoder(drop='first'), ['batting_team', 'bowling_team',
                                          'city'])], 
    remainder='passthrough'
)
```

```python
cf
```

# Creating the pipeline with LogisticRegression

```python
pipe = Pipeline(steps=[
    ('step1', cf),('step2', LogisticRegression(solver='liblinear') )])
```

```python
pipe
```

```python
# fitting the training data
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
# Reality Check : 0 (loose the match prob) - 46% and 1 (win probability)-53%
pipe.predict_proba(x_test)[10]
```

```python
pipe2 = Pipeline(steps=[
    ('step1', cf),('step2', RandomForestClassifier() )])
pipe2.fit(x_train, y_train)
print(metrics.accuracy_score(y_test, pipe2.predict(x_test)))
```

```python
print(metrics.classification_report(y_test, pipe2.predict(x_test)))
```
